```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Maneuvering TA

```python
df = pd.read_csv('physical_tests/maneuvering1/maneuvering1/actuators/actuator_1/thruster/motor_state.csv')

df['time'] = pd.to_datetime(df['time'])
df = df.sort_values('time').reset_index(drop=True)
start_time = df['time'].iloc[0]
df['seconds'] = (df['time'] - start_time).dt.total_seconds()

# Identify outliers using IQR method for voltage
v_q1 = df['battery_voltage'].quantile(0.25)
v_q3 = df['battery_voltage'].quantile(0.75)
v_iqr = v_q3 - v_q1
v_lower, v_upper = v_q1 - 1.5 * v_iqr, v_q3 + 1.5 * v_iqr
mask_volt = df['battery_voltage'].between(v_lower, v_upper)

# Identify outliers using IQR method for current
c_q1 = df['battery_current'].quantile(0.25)
c_q3 = df['battery_current'].quantile(0.75)
c_iqr = c_q3 - c_q1
c_lower, c_upper = c_q1 - 1.5 * c_iqr, c_q3 + 1.5 * c_iqr
mask_curr = df['battery_current'].between(c_lower, c_upper)


# Plot Battery Voltage over time
plt.figure(figsize=(10, 5))
plt.plot(df['seconds'][mask_volt].to_numpy(), df['battery_voltage'][mask_volt].to_numpy())
plt.xlabel('Seconds since start')
plt.ylabel('Battery Voltage (V)')
plt.title('Battery Voltage over Time Actuator 1')
plt.grid()
plt.tight_layout()
plt.show()

# Plot Battery Current over time
plt.figure(figsize=(10, 5))
plt.plot(df['seconds'][mask_curr].to_numpy(), df['battery_current'][mask_curr].to_numpy())
plt.xlabel('Seconds since start')
plt.ylabel('Battery Current (A)')
plt.title('Battery Current over Time Actuator 1')
plt.grid()
plt.tight_layout()
plt.show()

df['battery_power'] = df['battery_voltage'] * df['battery_current']

# Filter out outliers in battery power using IQR method
p_q1 = df['battery_power'].quantile(0.25)
p_q3 = df['battery_power'].quantile(0.75)
p_iqr = p_q3 - p_q1
p_lower, p_upper = p_q1 - 1.5 * p_iqr, p_q3 + 1.5 * p_iqr
mask_power = df['battery_power'].between(p_lower, p_upper)

# Plot filtered Battery Power over time
plt.figure(figsize=(10, 5))
plt.plot(df['seconds'][mask_power].to_numpy(), df['battery_power'][mask_power].to_numpy())
plt.xlabel('Seconds since start')
plt.ylabel('Battery Power (W)')
plt.title('Battery Power over Time Actuator 1')
plt.grid()
plt.tight_layout()
plt.show()

# Plot Motor Speed over time
plt.figure(figsize=(10, 5))
plt.plot(df['seconds'].to_numpy(), df['motor_speed'].to_numpy())
plt.xlabel('Seconds since start')
plt.ylabel('Motor Speed (RPM)')
plt.title('Motor Speed over Time Actuator 1')
plt.grid()
plt.tight_layout()
plt.show()

az_df = pd.read_csv('physical_tests/maneuvering1/maneuvering1/actuators/actuator_1/azimuth_angle.csv')

az_df['time'] = pd.to_datetime(az_df['time'])
az_df = az_df.sort_values('time').reset_index(drop=True)
start_time = az_df['time'].iloc[0]
az_df['seconds'] = (az_df['time'] - start_time).dt.total_seconds()

# Plot Azimuth Angle over time
plt.figure(figsize=(10, 5))
plt.plot(az_df['seconds'].to_numpy(), az_df['data'].to_numpy())
plt.xlabel('Seconds since start')
plt.ylabel('Azimuth Angle ()')
plt.title('Azimuth Angle over Time Actuator 1')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
df1 = pd.read_csv('thruster_data_test/thruster_data_test/actuators/actuator_3/thruster/motor_state.csv')
df2 = pd.read_csv('thruster_data_test/thruster_data_test/actuators/actuator_4/thruster/motor_state.csv')

for df in (df1, df2):
    df['battery_power'] = df['battery_voltage'] * df['battery_current']
    df.loc[df['motor_speed']==0, 'battery_power'] = 0

# --- 2. IQR filter function & apply ---
def iqr_filter(s):
    Q1, Q3 = s.quantile([0.25,0.75])
    IQR = Q3-Q1
    return s.between(Q1-1.5*IQR, Q3+1.5*IQR)

mask1 = iqr_filter(df1['motor_speed']) & iqr_filter(df1['battery_power'])
mask2 = iqr_filter(df2['motor_speed']) & iqr_filter(df2['battery_power'])
df1f, df2f = df1[mask1], df2[mask2]

# --- 3. Combine ---
df = pd.concat([df1f, df2f], ignore_index=True)
x = df['motor_speed'].values
y = df['battery_power'].values

# --- 4a. General cubic through zero ---
A_cubic = np.vstack([x**3, x**2, x]).T
a3, a2, a1 = np.linalg.lstsq(A_cubic, y, rcond=None)[0]
print(f'General cubic:   P = {a3:.3e} + {a2:.3e} + {a1:.3e}')

# --- 4b. Physics-informed ( + ) ---
A_phys = np.vstack([x**3, x]).T
b3, b1 = np.linalg.lstsq(A_phys, y, rcond=None)[0]
print(f'Phys-informed:   P = {b3:.3e} + {b1:.3e}')

# --- 4c. Weighted LS (phys-inf, w=||) ---
w = np.abs(x)**2
Aw = A_phys * w[:,None]
yw = y * w
bw3, bw1 = np.linalg.lstsq(Aw, yw, rcond=None)[0]
print(f'Weighted phys-inf: P = {bw3:.3e} + {bw1:.3e}')

# --- 4d. Envelope fit at 95th percentile (phys-inf) ---
# Bin motor speeds
n_bins = 12
df['bin'] = pd.cut(df['motor_speed'], bins=n_bins)
env = df.groupby('bin')['battery_power'].quantile(0.95).dropna()
bin_edges = np.linspace(x.min(), x.max(), n_bins+1)
bin_centers = bin_edges[:-1] + np.diff(bin_edges)/2
x_env = bin_centers[~env.isna()]
y_env = env.values

Ae = np.vstack([x_env**3, x_env]).T
e3, e1 = np.linalg.lstsq(Ae, y_env, rcond=None)[0]
print(f'Envelope fit:    P = {e3:.3e} + {e1:.3e}')

# --- 5. Loglog power law fit ---
mask_pos = (x>0)&(y>0)
logx, logy = np.log(x[mask_pos]), np.log(y[mask_pos])
m, c = np.polyfit(logx, logy, 1)
print(f'Loglog fit:     P  exp({c:.3f})^{m:.3f}')

# --- 6. Plot them all ---
 = np.linspace(x.min(), x.max(), 400)
plt.figure(figsize=(10,6))
plt.scatter(x, y, s=20, alpha=0.4, label='Filtered data')

plt.plot(, a3***3 + a2***2 + a1*, 'C0-', lw=2, label='General cubic')
plt.plot(, b3***3 + b1*,       'C1--',lw=2, label='Phys-informed +')
plt.plot(, bw3***3 + bw1*,    'C2-.',lw=2, label='Weighted +')
plt.plot(, e3***3 + e1*,      'C3:', lw=2, label='95% envelope +')
plt.plot(, np.exp(c)***m,      'k:',   lw=2, label='Loglog power law')

plt.xlabel('Motor Speed ')
plt.ylabel('Battery Power P [W]')
plt.title('Thruster Power Fits (All Through Origin)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
