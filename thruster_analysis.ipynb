```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Maneuvering TA

```python
df = pd.read_csv('physical_tests/maneuvering1/maneuvering1/actuators/actuator_1/thruster/motor_state.csv')

df['time'] = pd.to_datetime(df['time'])
df = df.sort_values('time').reset_index(drop=True)
start_time = df['time'].iloc[0]
df['seconds'] = (df['time'] - start_time).dt.total_seconds()

# Identify outliers using IQR method for voltage
v_q1 = df['battery_voltage'].quantile(0.25)
v_q3 = df['battery_voltage'].quantile(0.75)
v_iqr = v_q3 - v_q1
v_lower, v_upper = v_q1 - 1.5 * v_iqr, v_q3 + 1.5 * v_iqr
mask_volt = df['battery_voltage'].between(v_lower, v_upper)

# Identify outliers using IQR method for current
c_q1 = df['battery_current'].quantile(0.25)
c_q3 = df['battery_current'].quantile(0.75)
c_iqr = c_q3 - c_q1
c_lower, c_upper = c_q1 - 1.5 * c_iqr, c_q3 + 1.5 * c_iqr
mask_curr = df['battery_current'].between(c_lower, c_upper)


# Plot Battery Voltage over time
plt.figure(figsize=(10, 5))
plt.plot(df['seconds'][mask_volt].to_numpy(), df['battery_voltage'][mask_volt].to_numpy())
plt.xlabel('Seconds since start')
plt.ylabel('Battery Voltage (V)')
plt.title('Battery Voltage over Time Actuator 1')
plt.grid()
plt.tight_layout()
plt.show()

# Plot Battery Current over time
plt.figure(figsize=(10, 5))
plt.plot(df['seconds'][mask_curr].to_numpy(), df['battery_current'][mask_curr].to_numpy())
plt.xlabel('Seconds since start')
plt.ylabel('Battery Current (A)')
plt.title('Battery Current over Time Actuator 1')
plt.grid()
plt.tight_layout()
plt.show()

df['battery_power'] = df['battery_voltage'] * df['battery_current']

# Filter out outliers in battery power using IQR method
p_q1 = df['battery_power'].quantile(0.25)
p_q3 = df['battery_power'].quantile(0.75)
p_iqr = p_q3 - p_q1
p_lower, p_upper = p_q1 - 1.5 * p_iqr, p_q3 + 1.5 * p_iqr
mask_power = df['battery_power'].between(p_lower, p_upper)

# Plot filtered Battery Power over time
plt.figure(figsize=(10, 5))
plt.plot(df['seconds'][mask_power].to_numpy(), df['battery_power'][mask_power].to_numpy())
plt.xlabel('Seconds since start')
plt.ylabel('Battery Power (W)')
plt.title('Battery Power over Time Actuator 1')
plt.grid()
plt.tight_layout()
plt.show()

# Plot Motor Speed over time
plt.figure(figsize=(10, 5))
plt.plot(df['seconds'].to_numpy(), df['motor_speed'].to_numpy())
plt.xlabel('Seconds since start')
plt.ylabel('Motor Speed (RPM)')
plt.title('Motor Speed over Time Actuator 1')
plt.grid()
plt.tight_layout()
plt.show()

az_df = pd.read_csv('physical_tests/maneuvering1/maneuvering1/actuators/actuator_1/azimuth_angle.csv')

az_df['time'] = pd.to_datetime(az_df['time'])
az_df = az_df.sort_values('time').reset_index(drop=True)
start_time = az_df['time'].iloc[0]
az_df['seconds'] = (az_df['time'] - start_time).dt.total_seconds()

# Plot Azimuth Angle over time
plt.figure(figsize=(10, 5))
plt.plot(az_df['seconds'].to_numpy(), az_df['data'].to_numpy())
plt.xlabel('Seconds since start')
plt.ylabel('Azimuth Angle (°)')
plt.title('Azimuth Angle over Time Actuator 1')
plt.grid()
plt.tight_layout()
plt.show()
```
